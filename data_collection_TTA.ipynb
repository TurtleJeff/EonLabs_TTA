```python
import pandas as pd                        
from pytrends.request import TrendReq

from pathlib import Path
from datetime import date
from calendar import monthrange
from time import sleep
```

```python
# settings
output_folder = Path('X:/EonLabs_TTA/output')
geo_location = '' # DEFAULT - WORLDWIDE, 'US' - US, 'US-AL' - Alabama
kw_list = ['bitcoin']

# connect to google
pytrends = TrendReq(hl='en-US', tz=360, backoff_factor=1) # tz - timezone offset
today = date.today()
```

```python
# Daily Data - retrive data of each month and concat
def get_daily_data(year_start, month_start, year_end, month_end, pytrends=pytrends, geo_location=geo_location):
    # 1.generate (year, month) pairs
    year_month_pairs = []
    curr_year, curr_month = year_start, month_start
    
    while curr_year < year_end or curr_month <= month_end:
        year_month_pairs.append((curr_year, curr_month))
    
        if curr_month == 12:
            curr_year += 1
            curr_month = 1
        else:
            curr_month += 1
    
    # 2.retrive the data
    all_dataframes = []
    idx = 0
    while idx < len(year_month_pairs):
        year, month = year_month_pairs[idx]
        print(f'working on year: {year}, month: {month}')

        first_day = f'{year}-{month}-01'
        last_day = f'{year}-{str(month).zfill(2)}-{monthrange(year, month)[1]}'
        try:
            pytrends.build_payload(kw_list, cat=0, timeframe=first_day+' '+last_day, geo=geo_location, gprop='')
            all_dataframes.append(pytrends.interest_over_time())
        except:
            idx -= 1
            print('Request Error, wait for another minute...')
            sleep(60)
        sleep(2)
        idx += 1
    return pd.concat(all_dataframes)
```

```python
daily_df = get_daily_data(year_start=2015, month_start=1, year_end=2022, month_end=9)
daily_df.to_csv(output_folder/'my_daily_data.csv')
```

```python
# Weekly Data - retrive data of each year and concat
def get_weekly_data(year_start, year_end, pytrends=pytrends, geo_location=geo_location):
    all_dataframes = []
    for curr_year in range(year_start, year_end+1):
        print(f'working on year: {curr_year}')
        first_day = f'{curr_year}-01-01'
        last_day = f'{curr_year}-12-31'
        pytrends.build_payload(kw_list, cat=0, timeframe=first_day+' '+last_day, geo=geo_location, gprop='')
        all_dataframes.append(pytrends.interest_over_time())
        
    return pd.concat(all_dataframes)
```

```python
weekly_df = get_weekly_data(year_start=2015, year_end=2022)
weekly_df.to_csv(output_folder/'my_weekly_data.csv')
```
